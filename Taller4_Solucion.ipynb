```python
import re
import pandas as pd
from collections import Counter
import numpy as np

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("C:/Users/Danilo P/Documents/KonradLorenz/NLP/Taller4/dialogos.csv")
df
```

```python
pre = [re.sub(r"[\W\d]"," ",a.lower()) for a in df['Locución']]
```

```python
pre2 = []
for a in pre:
    texto = a.strip()
    pre2.append(texto.split())
```

```python
resultado = []
for x in pre2:
    result = [pala for pala in x if pala not in stopwords_sp]
    resultado.append(' '.join([str(elem) for elem in result]))
```

```python
df['pre-procesado']=resultado
df
```

```python
lista_conteo = []
for w in resultado:
    y = w.split()
    for p in y:
        lista_conteo.append(p)
lista_conteo
```

```python
Counter(lista_conteo).most_common(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
bow_rep = count_vect.fit_transform(df["pre-procesado"].values)

bowfinal = bow_rep.toarray().tolist()
```

```python
df['bow'] = bowfinal
df
```
