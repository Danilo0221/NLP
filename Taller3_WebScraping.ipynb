```python
import bs4 as bs
import urllib.request
import re
```

```python
links = ['https://es.wikipedia.org/wiki/Panthera_leo', 'https://es.wikipedia.org/wiki/Giraffa_camelopardalis',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus', 'https://es.wikipedia.org/wiki/Pantera_negra',
        'https://es.wikipedia.org/wiki/Ursidae','https://es.wikipedia.org/wiki/Selachimorpha',
        'https://es.wikipedia.org/wiki/Eunectes', 'https://es.wikipedia.org/wiki/Chamaeleonidae',
        'https://es.wikipedia.org/wiki/Colibr%C3%AD_(ave)', 'https://es.wikipedia.org/wiki/%C3%81guila']
```

```python
#Hacer Web Scraping de 10 animales en Wikipedia (en búcle)
soup = []
for x in links:
    req = urllib.request.Request(x, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urllib.request.urlopen(req)
    source = webpage.read()
    webpage.close()
    soup.append(bs.BeautifulSoup(source, 'html.parser'))
```

```python
print(soup)
```

```python
#Obtener el encabezado de cada animal
for y in range(0,len(soup)):
    print("Encabezado:", soup[y].find("h1").text)
```
